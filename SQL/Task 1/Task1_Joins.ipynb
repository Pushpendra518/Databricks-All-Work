```sql
create schema joins.tasks;
```

```sql
use joins.tasks;
```

```sql
-- Locations table
create table if not exists joins.tasks.Locations(
  LocationID int primary key,
  city string,
  Country string
)
using delta;
```

```sql
-- Departments table
create table if not exists joins.tasks.Departments(
  DeptID int primary key,
  DeptName string,
  LocationID int,
  Budget decimal(10,2),
  foreign key(LocationID) references Locations(LocationID)
)
using delta;
```

```sql
-- Employees table
create table if not exists joins.tasks.Employees(
  EmployeeID int primary key,
  FirstName string,
  LastName string,
  DeptID int,
  ManagerID int,
  Salary decimal(10,2),
  HireDate date,
  foreign key(DeptID) references Departments(DeptID),
  foreign key(ManagerID) references Employees(EmployeeID)
)
using delta;
```

```sql
-- Projects table
create table if not exists joins.tasks.Projects(
  ProjectID int primary key,
  ProjectName string,
  StartDate date,
  EndDate date,
  DeptID int,
  foreign key(DeptID) references Departments(DeptID)
)
using delta;

-- Project_Assignments
create table if not exists joins.tasks.Project_Assignments(
  ProjectID int,
  EmployeeID int,
  Role string,
  foreign key(ProjectID) references Projects(ProjectID),
  foreign key(EmployeeID) references Employees(EmployeeID),
  primary key(ProjectID, EmployeeID)
)
using delta;
```

```sql
-- Salary_Grades table
create table if not exists joins.tasks.Salary_Grades(
 Grade char(1) primary key,
  MinSalary decimal(10,2),
  MaxSalary decimal(10,2)
)
using delta;
```

**Q1. Staff Directory: Generate a list of all employees and the name of the department they work in.**

```sql
select e.FirstName, e.LastName, d.DeptName 
from joins.tasks.employees e join 
joins.tasks.departments d
on e.DeptID = d.DeptID
```

**Q2. Regional Breakdown: List all department names and the city they are located in.**

```sql
select d.DeptName, l.City, l.Country 
from joins.tasks.departments d join
joins.tasks.locations l
on d.LocationID = l.LocationID
```

**Q3. Project Ownership: Display every project name along with the name of the
department responsible for it.**

```sql
select p.ProjectName, d.DeptName
from joins.tasks.projects p join
joins.tasks.departments d
on p.DeptID = d.DeptID
```

**Q4. Grade Assignment: Show each employee's full name, their salary, and the
corresponding salary grade (A, B, or C) they fall into.**

```sql
select concat(e.FirstName, ' ', e.LastName) as FullName, e.Salary, s.Grade
from joins.tasks.employees e 
join joins.tasks.salary_grades s
on e.Salary between s. MinSalary and s.MaxSalary
order by e.Salary desc
```

**Q5. Role Identification: List all employees and the specific roles they are playing in their
assigned projects.**

```sql
select e.FirstName, e.LastName,
p.ProjectName,
pa.Role
from joins.tasks.employees e
join joins.tasks.project_assignments pa
on e.EmployeeID = pa.EmployeeID
join joins.tasks.projects p
on pa.ProjectID = p.ProjectID
```

**Q12. Project Manpower: Show a list of all projects and a count of how many employees
are assigned to each.**

```sql
select pa.ProjectID, p.ProjectName, count(e.EmployeeID) as total_member
from joins.tasks.project_assignments pa
left join joins.tasks.employees e
on pa.EmployeeID = e.EmployeeID
join joins.tasks.projects p
on pa.ProjectID = p.ProjectID
group by pa.ProjectID, p.ProjectName
```

**Q13. Location Headcount: Display each city and the total number of employees working
at that location.**

```sql
select l.City , count(e.EmployeeID)
from joins.tasks.locations l
left join joins.tasks.departments d
on l.LocationID = d.LocationID
left join joins.tasks.employees e
on d.DeptID = e.DeptID
group by l.City
```

**Q14. Over-Budget Check: Find employees who earn more than the average salary of the
"Sales" department (by name).**

```sql
select concat(e.FirstName, ' ',e.LastName) as FullName, e.Salary
from joins.tasks.employees e
join joins.tasks.departments d
on d.DeptID = e.DeptID
where e.Salary > (
    select avg(e1.Salary)
    from joins.tasks.employees e1
    left join joins.tasks.departments d1
    on d1.DeptID = e1.DeptID
    where d1.DeptName = 'Sales'
)
```

**Q15. Grade Distribution: Count how many employees fall into each Salary Grade.**

```sql
select s.Grade , count(e.EmployeeID)
from joins.tasks.salary_grades s
left join joins.tasks.employees e
on e.Salary between s.MinSalary and s.MaxSalary
group by s.Grade
order by count(e.EmployeeID) desc
```

**Q19. Managerial Oversight: List all managers and the number of subordinates they are

**Q20. Double-Department Match: Find projects that are being handled by a department
located in 'Udaipur'.**

```sql
select l.City ,d.DeptName ,p.ProjectName
from joins.tasks.locations l
join joins.tasks.departments d
on l.LocationID = d.LocationID
join joins.tasks.projects p 
on d.DeptID = p.DeptID
where l.City = 'Udaipur'
```

**Q21. The Full Stack: Display Employee Name, Department Name, Project Name, and the
City where the department is based.**

```sql
select concat(e.FirstName, ' ', e.LastName) as FullName, d.DeptName , p.ProjectName, l.City
       from joins.tasks.employees e
       join joins.tasks.departments d  
       on e.DeptID = d.DeptID
       join joins.tasks.locations l
       on l.LocationID = d.LocationID 
       join joins.tasks.project_assignments pa
       on pa.EmployeeID = e.EmployeeID
       join joins.tasks.projects p 
       on pa.ProjectID = p.ProjectID
```

**Q22. Cross-Department Projects: List employees who are working on projects that
belong to a department other than their own.**

```sql
--select * from joins.tasks.departments
```

```sql
--    employee - project_assign - project
--        |                           |
--   department d1               department d2

select concat(e.FirstName, ' ', e.LastName) as FullName,
       d1.DeptName,
       p.ProjectName,
       d2.DeptName as projectDepartment
       from joins.tasks.employees e 
       join joins.tasks.project_assignments pa
       on e.EmployeeID = pa.EmployeeID
       join joins.tasks.projects p 
       on pa.ProjectID = p.ProjectID
       join joins.tasks.departments d1
       on e.DeptID = d1.DeptID
       join joins.tasks.departments d2
       on p.DeptID = d2.DeptID
       where e.DeptID <> p.DeptID
```

**Q23. Regional Project Count: Show a list of countries and the total number of active
projects being managed in each country.**

```sql
select l.Country, count(p.ProjectID) as totalProject
from joins.tasks.locations l
join joins.tasks.departments d
on l.LocationID = d.LocationID
join joins.tasks.projects p
on d.DeptID = p.DeptID
group by l.Country
order by totalProject desc
```

**Q24. Salary Grade by Location: Find the average salary for "Grade A" employees in each
specific city.**

```sql
select l.City,
       avg(e.Salary) as avgSalary,
       s.Grade
       from joins.tasks.locations l
       join joins.tasks.departments d 
       on l.LocationID = d.LocationID
       join joins.tasks.employees e  
       on d.DeptID = e.DeptID
       join joins.tasks.salary_grades s
       on e.Salary between s.MinSalary and s.MaxSalary
       where s.Grade = 'A'
       group by l.city, s.Grade
```

**Q25. The Master Audit: Generate a report showing all Departments, their Projects, and
the Lead names (Role = 'Lead') assigned to those projects, ensuring departments
without projects are still listed.**

```sql
select d.DeptName,
       p.ProjectName,
       concat(e.FirstName, ' ', e.LastName) as Lead
       from joins.tasks.departments d
       left join joins.tasks.projects p
       on d.DeptID = p.DeptID
       left join joins.tasks.project_assignments pa
       on p.ProjectID = pa.ProjectID
       and role = 'Lead'
       left join joins.tasks.employees e
       on pa.EmployeeID = e.EmployeeID
```
